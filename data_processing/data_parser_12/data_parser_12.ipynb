```python
dataset = pd.read_excel("ОБ1.xlsx")
```

перебиваем названия

```python
names_target = [
    "Наименование статей расходов",
    "Категория",
]
key_column = "Наименование подстатей расходов"
names_dict = {}
for i in range(len(dataset_categories)):
    line = dataset_categories.iloc[i]
    dataset_key_col = line[key_column]
    names_dict[dataset_key_col] = {}
    for col_name in names_target:
        names_dict[dataset_key_col] = {
            **names_dict[dataset_key_col],
            col_name: line[col_name],
        }

for i in range(len(dataset)):
    line = dataset.iloc[i]
    dataset_key_col = line[key_column]
    if dataset_key_col in names_dict:
        # print(line)
        for col_name in names_target:
            dataset[col_name].iloc[i] = names_dict[dataset_key_col][col_name]
    # break
```

сохраняем


```python
dataset.to_excel(
    "./ob1_test.xlsx",
    index=False,
)
```

смотрим отличия, выводим номера строк с отличием


```python
new_dataset = pd.read_excel("./ob1_test.xlsx")
dataset = pd.read_excel("ОБ1.xlsx")

for i in range(len(dataset)):
    line = dataset.iloc[i]
    line_2 = new_dataset.iloc[i]
    dataset_key_col = line[key_column]
    if dataset_key_col in names_dict:
        # print(line)
        for col_name in names_target:
            if line[col_name] != line_2[col_name]:
                print(i + 1)
```
